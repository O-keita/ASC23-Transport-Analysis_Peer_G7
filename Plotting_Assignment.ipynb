```python
# YOUR CODE HERE
# Histogram 1 (co2)
plt.figure(figsize=(8, 6))
plt.hist(data['co2'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of carbon Dioxide')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# YOUR CODE HERE
# Histogram 2 (comb)
plt.figure(figsize=(8, 6))
plt.hist(data['comb'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Combined Fuel Efficiency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# YOUR CODE HERE
# Histogram 3 (displ)
plt.figure(figsize=(8, 6))
plt.hist(data['displ'], bins=20, color='skyblue', edgecolor='black')
plt.title('Engine displacement')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
